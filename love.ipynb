```python
for love in range(2000, 9999):
    l, o, v, e = love//1000, love//100 % 10, love//10 % 10, love % 10
    s, o_, l_, v_, e_ = list(map(int, str(5 * love)))
    if len(set([s, o_, l_, v_, e_])) == 5 and [s, o_, l_, v_, e_] == [s, o, l, v, e] :
        print("LOVE", "=", "".join([str(love)]))
```
